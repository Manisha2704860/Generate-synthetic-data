# Task
Generate synthetic data for a rare event using GANs or VAEs, given a small dataset of the rare event.

## Data preparation

### Subtask:
Load and preprocess your small dataset. This might involve cleaning, normalizing, and structuring the data for use with generative models.


**Reasoning**:
Load the data into a pandas DataFrame and perform initial preprocessing steps including handling missing values and scaling numerical features.



```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assume the small dataset is in a CSV file named 'rare_event_data.csv'
# and it has numerical and potentially categorical features.
try:
    df = pd.read_csv('rare_event_data.csv')
except FileNotFoundError:
    # Create a dummy dataset if the file is not found for demonstration
    data = {
        'numerical_feature_1': [1.5, 2.1, 1.8, 2.5, 1.9, 10.0, 2.0, 2.2, 1.6, 2.3],
        'numerical_feature_2': [100, 105, 98, 110, 102, 500, 103, 106, 99, 108],
        'categorical_feature': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'A', 'B', 'C']
    }
    df = pd.DataFrame(data)
    df.to_csv('rare_event_data.csv', index=False)
    print("Created a dummy 'rare_event_data.csv' for demonstration.")

# Handle missing values (example: fill with mean for numerical, mode for categorical)
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Normalize numerical features
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Convert categorical features using one-hot encoding
df = pd.get_dummies(df, columns=['categorical_feature'], drop_first=True)

# Display the preprocessed data
display(df.head())
```

## Choose a generative model

### Subtask:
Select an appropriate model like a GAN or VAE based on the type and complexity of your data.


**Reasoning**:
Review the data in the `df` DataFrame and select an appropriate generative model (GAN or VAE) based on its characteristics.



```python
print("DataFrame Info:")
df.info()
print("\nDataFrame Description:")
display(df.describe())

# Based on the small size and tabular nature of the data, a VAE might be a more stable option
# compared to GANs, which can be challenging to train on small, structured datasets.
# VAEs are also generally easier to implement and debug.
chosen_model = "VAE"
print(f"\nChosen generative model: {chosen_model}")
```

## Implement and train the model

### Subtask:
Write code to build and train the chosen generative model on your small dataset.


**Reasoning**:
Define, implement, and train the VAE model on the preprocessed data. This involves defining the encoder and decoder architectures, implementing the VAE class with the reparameterization trick and loss function, and setting up and running the training loop.



**Reasoning**:
The traceback indicates an `AttributeError` because `keras.losses.mean_squared_error` is not the correct way to access the mean squared error loss function in the current Keras version. The correct way is likely `tf.keras.losses.MeanSquaredError()`. I will fix this in the `train_step` method of the VAE class.



**Reasoning**:
The traceback shows another `AttributeError`, this time because `tf.gradienttape` is incorrect. The correct way to access the Gradient Tape is `tf.GradientTape()`. I need to fix this in the `train_step` method.



```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Get the input dimension from the preprocessed dataframe
input_dim = df.shape[1]
latent_dim = 2  # Define the dimensionality of the latent space

# Define the VAE model
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        # Instantiate the loss function here
        self.mean_squared_error = tf.keras.losses.MeanSquaredError()


    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        reconstruction = self.decoder(z)
        return reconstruction

    def train_step(self, data):
        with tf.GradientTape() as tape: # Corrected GradientTape usage
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)
            # Use the instantiated loss function
            reconstruction_loss = tf.reduce_mean(
                self.mean_squared_error(data, reconstruction)
            )
            kl_loss = -0.5 * tf.reduce_mean(
                z_log_var - tf.square(z_mean) - tf.exp(z_log_var) + 1
            )
            total_loss = reconstruction_loss + kl_loss
        grads = tape.gradient(total_loss, self.trainable_variables) # Corrected gradient call
        self.optimizer.apply_gradients(zip(grads, self.trainable_variables))
        return {
            "loss": total_loss,
            "reconstruction_loss": reconstruction_loss,
            "kl_loss": kl_loss,
        }

# Define the encoder
encoder_inputs = keras.Input(shape=(input_dim,))
x = layers.Dense(64, activation="relu")(encoder_inputs)
x = layers.Dense(32, activation="relu")(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.keras.backend.random_normal(shape=(tf.shape(z_mean)[0], latent_dim))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = layers.Lambda(sampling)([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# Define the decoder
decoder_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(32, activation="relu")(decoder_inputs)
x = layers.Dense(64, activation="relu")(x)
decoder_outputs = layers.Dense(input_dim, activation="sigmoid")(x) # Sigmoid for output between 0 and 1
decoder = keras.Model(decoder_inputs, decoder_outputs, name="decoder")

# Instantiate and compile the VAE
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam())

# Train the VAE
# Convert dataframe to tensor
data_tensor = tf.constant(df.values, dtype=tf.float32)

epochs = 100 # Define number of epochs
batch_size = 8 # Define batch size

history = vae.fit(data_tensor, epochs=epochs, batch_size=batch_size)

print("VAE training finished.")
```

## Generate synthetic data

### Subtask:
Use the trained model to generate new, synthetic data points that resemble the rare event.


**Reasoning**:
Generate synthetic data points using the trained VAE decoder, convert them to a DataFrame, and display the head of the synthetic data.



```python
import pandas as pd
import tensorflow as tf

# Generate random points in the latent space
num_synthetic_samples = 50  # Define the number of synthetic data points to generate
random_latent_vectors = tf.random.normal(shape=(num_synthetic_samples, latent_dim))

# Use the decoder to generate synthetic data
synthetic_data_tensor = vae.decoder(random_latent_vectors)

# Convert the synthetic data tensor to a pandas DataFrame
synthetic_df = pd.DataFrame(synthetic_data_tensor.numpy(), columns=df.columns)

# Display the head of the synthetic DataFrame
display(synthetic_df.head())
```

## Evaluate the synthetic data

### Subtask:
Assess the quality and usefulness of the generated data. This could involve visualizing the data, comparing its statistical properties to the real data, or using it to train a downstream model and evaluating its performance.


**Reasoning**:
Visualize the distributions of numerical and categorical features in both the original and synthetic data using histograms and bar plots respectively, and calculate descriptive statistics for numerical features to assess the quality of the generated data.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize numerical features
numerical_cols = df.select_dtypes(include=['float64']).columns

plt.figure(figsize=(12, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, 2, i + 1)
    sns.histplot(df[col], kde=True, color='skyblue', label='Original')
    sns.histplot(synthetic_df[col], kde=True, color='lightcoral', label='Synthetic')
    plt.title(f'Distribution of {col}')
    plt.legend()
plt.tight_layout()
plt.show()

# Compare statistical properties of numerical features
print("Original Data Numerical Feature Statistics:")
display(df[numerical_cols].describe())

print("\nSynthetic Data Numerical Feature Statistics:")
display(synthetic_df[numerical_cols].describe())

# Visualize categorical features
# Apply a threshold to synthetic categorical features for visualization
categorical_cols = df.select_dtypes(include=['bool']).columns
synthetic_df_thresholded = synthetic_df.copy()
for col in categorical_cols:
    synthetic_df_thresholded[col] = (synthetic_df_thresholded[col] > 0.5).astype(bool)


plt.figure(figsize=(12, 5))
for i, col in enumerate(categorical_cols):
    plt.subplot(1, 2, i + 1)
    df[col].value_counts().plot(kind='bar', color='skyblue', alpha=0.7, label='Original')
    synthetic_df_thresholded[col].value_counts().plot(kind='bar', color='lightcoral', alpha=0.7, label='Synthetic')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=0)
    plt.legend()
plt.tight_layout()
plt.show()
```

## Iterate and refine

### Subtask:
Based on the evaluation, refine your data preparation, model choice, or training process to improve the quality of the synthetic data.


**Reasoning**:
Based on the evaluation in the previous step, the distributions of the synthetic data, especially for the categorical features after thresholding, do not perfectly match the original data. This suggests that the current VAE architecture or training process may not be optimally capturing the underlying data distribution. To improve the quality of the synthetic data, I will refine the VAE model by adding more layers to the encoder and decoder to potentially increase its capacity to learn complex patterns. I will also increase the number of epochs to allow the model more training time to converge.



```python
# Refine the VAE model architecture and training process

# Define a more complex encoder
encoder_inputs = keras.Input(shape=(input_dim,))
x = layers.Dense(128, activation="relu")(encoder_inputs) # Increased neurons
x = layers.Dense(64, activation="relu")(x) # Added another layer
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = tf.keras.backend.random_normal(shape=(tf.shape(z_mean)[0], latent_dim))
    return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = layers.Lambda(sampling)([z_mean, z_log_var])
encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# Define a more complex decoder
decoder_inputs = keras.Input(shape=(latent_dim,))
x = layers.Dense(64, activation="relu")(decoder_inputs) # Added another layer
x = layers.Dense(128, activation="relu")(x) # Increased neurons
decoder_outputs = layers.Dense(input_dim, activation="sigmoid")(x) # Sigmoid for output between 0 and 1
decoder = keras.Model(decoder_inputs, decoder_outputs, name="decoder")

# Instantiate and compile the VAE with the updated architecture
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam())

# Increase the number of epochs for training
epochs = 200 # Increased epochs
batch_size = 8 # Keep the same batch size

# Convert dataframe to tensor
data_tensor = tf.constant(df.values, dtype=tf.float32)

print(f"Training VAE with {epochs} epochs and a more complex architecture...")
history = vae.fit(data_tensor, epochs=epochs, batch_size=batch_size)

print("Refined VAE training finished.")
```

The VAE model has been refined and retrained. Now, I need to generate synthetic data using this refined model to see if the changes have improved the data quality. I will use the same process as before: generate random points in the latent space and use the decoder to generate synthetic data, then convert it to a pandas DataFrame.



```python
# Generate synthetic data using the refined VAE model

# Generate random points in the latent space
# Use the same number of synthetic samples as before
random_latent_vectors = tf.random.normal(shape=(num_synthetic_samples, latent_dim))

# Use the refined decoder to generate synthetic data
synthetic_data_tensor_refined = vae.decoder(random_latent_vectors)

# Convert the synthetic data tensor to a pandas DataFrame
synthetic_df_refined = pd.DataFrame(synthetic_data_tensor_refined.numpy(), columns=df.columns)

# Display the head of the refined synthetic DataFrame
display(synthetic_df_refined.head())
```
